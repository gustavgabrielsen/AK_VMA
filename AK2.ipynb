## 1. The Maastricht Treaty and Fiscal Discipline
From,

$b(t) = (r - g)b(t) - s, \hspace{1cm}$ r,g,s are constants

We can find steady-state $b*$ when $b=0$ which gives;

$b = \frac{s}{r - g}$

From the expression we can see that if $r - g > 0$ it is asymptotically unstable. And if $r - g < 0$ it is asymptotically stable.

#### a)
$r = 3\%$, $g = 2\%$, $s = 0.5\%$

$$
r - g = 0.01 > 0 \rightarrow \text{unstable equilibruim}
$$

$$
b* = \frac{0.05}{0.01} = 0.5 = 50\%
$$

**Scenario 1**: b(0) = 70%

Since the interest rate exceeds the growth rate, any debt levels above the steady state ($b(0) > b*$), will lead to an increased debt ratio over time. 
The country moves further away from the steady state target instead of converging towards it.

**Scenario 2**: b(0) = 55%

Similar to scenario 1, the debt ratio will diverge away from the steady state over time. 

As prime minister, I would implement policies to increase the steady state level, such as cutting government spending or increasing taxes to generate budget surpluses. This could change the primary surplus $s$ to a higher value, which would help stabilize the debt ratio over time. If we would want to lower the debt ratio we could by increasing $s$ to $0.6\%$, then the 60% rule would be put in motion, and although the debt ratio still would diverge, it would diverge downwards for **Scenario 2**. For **Scenario 1** we would need an even more aggresive fiscal policy making $s\ge 0.7\%$ to either stabilize or lower the debt ratio.

### b)
$r = 1\%$, $g = 3\%$, $s = - 1.6\%$

$$
r - g = -0.02 < 0 \rightarrow \text{stable equilibruim}
$$

$$
b* = \frac{-0.016}{-0.02} = 0.8 = 80\%
$$

Since the growth rate exceeds the interest rate, any debt levels above the steady state ($b(0) > b*$), or below the steady state ($b(0) < b*$), will over time return to the steady state of 80%. Over time the debt ratio would increase, not staying at 60\% but converging towards 80\%. If we wanted to aim for the 60% rule, I would implement policies to decrease primary surplus spending $s$ to $s=-0.12\%$ which would lower the steady state equilibrium to 60%.



## 2. Managing the Stock of a Renewable Resource

We have the following differential equation modeling the fish population:

$$
\dot{p}(t) = r(p(t)) - (\mu + nk)p(t)
$$

### (a)

Steady state is found by setting $ p(t) = p_{ss} $ and $ \dot{p}(t) = 0 $:

$$
0 = \sqrt{p_{ss}} - (\mu + nk)p_{ss}
$$

$$
\sqrt{p_{ss}} = (\mu + nk)p_{ss}
$$

$$
1 = (\mu + nk)\sqrt{p_{ss}}
$$

$$
\boxed{p_{ss} = \frac{1}{(\mu + nk)^2}}
$$

### (b)

If $f(p) = \sqrt{p} - (\mu + nk)p$, the stability of the steady state can be analyzed by the derivative:

$$
f'(p) = \frac{1}{2\sqrt{p}} - (\mu + nk)
$$

Steady state:

$$
\frac{1}{\sqrt{p_{ss}}} = (\mu + nk)
$$

$$
f'(p_{ss}) = \frac{1}{2}(\mu + nk) - (\mu + nk) = -\frac{1}{2}(\mu + nk) < 0
$$

The steady state is negative therefore is is asymptotically stable.

### (c)

Given:

$$
\mu = 0.03, \quad n = 20, \quad k = 0.001
$$

$$
\mu + nk = 0.03 + 20(0.001) = 0.05
$$

$$
p_{ss} = \frac{1}{0.05^2} = 400
$$

Steady state fish population:

$$
\boxed{p_{ss} = 400}
$$

### (d)

Now the death rate increases, and the fishing stops:
$$
\mu = 0.06, \quad n = 0
$$

$$
\mu + nk = 0.06
$$

$$
p_{ss} = \frac{1}{0.06^2} = 277.78
$$

Comparison:

| Scenario | $\mu$ | $n$ | $p_{ss}$ |
|-----------|-----------|---------|---------------|
| (c) | 0.03 | 20 | 400 |
| (d) | 0.06 | 0 | 277.8 |

Conclusion:

The steady state fish population decreases from 400 to approximately 277.8 when the natural death rate increases and fishing stops.
