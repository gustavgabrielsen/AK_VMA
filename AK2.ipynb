## 2. Managing the Stock of a Renewable Resource

We have the following differential equation modeling the fish population:

$$
\dot{p}(t) = r(p(t)) - (\mu + nk)p(t)
$$

### (a)

Steady state is found by setting $ p(t) = p_{ss} $ and $ \dot{p}(t) = 0 $:

$$
0 = \sqrt{p_{ss}} - (\mu + nk)p_{ss}
$$

$$
\sqrt{p_{ss}} = (\mu + nk)p_{ss}
$$

$$
1 = (\mu + nk)\sqrt{p_{ss}}
$$

$$
\boxed{p_{ss} = \frac{1}{(\mu + nk)^2}}
$$

### (b)

If $f(p) = \sqrt{p} - (\mu + nk)p$, the stability of the steady state can be analyzed by the derivative:

$$
f'(p) = \frac{1}{2\sqrt{p}} - (\mu + nk)
$$

Steady state:

$$
\frac{1}{\sqrt{p_{ss}}} = (\mu + nk)
$$

$$
f'(p_{ss}) = \frac{1}{2}(\mu + nk) - (\mu + nk) = -\frac{1}{2}(\mu + nk) < 0
$$

The steady state is negative therefore is is asymptotically stable.

### (c)

Given:

$$
\mu = 0.03, \quad n = 20, \quad k = 0.001
$$

$$
\mu + nk = 0.03 + 20(0.001) = 0.05
$$

$$
p_{ss} = \frac{1}{0.05^2} = 400
$$

Steady state fish population:

$$
\boxed{p_{ss} = 400}
$$

### (d)

Now the death rate increases, and the fishing stops:
$$
\mu = 0.06, \quad n = 0
$$

$$
\mu + nk = 0.06
$$

$$
p_{ss} = \frac{1}{0.06^2} = 277.78
$$

Comparison:

| Scenario | $\mu$ | $n$ | $p_{ss}$ |
|-----------|-----------|---------|---------------|
| (c) | 0.03 | 20 | 400 |
| (d) | 0.06 | 0 | 277.8 |

Conclusion:

The steady state fish population decreases from 400 to approximately 277.8 when the natural death rate increases and fishing stops.
