## Oppgave 1 – The Maastricht Treaty and Fiscal Discipline

Vi starter fra differensialligningen:

$$
\dot{b}(t) = (r - g)b(t) - s
$$

hvor  
- ( b(t)): offentlig gjeld som andel av BNP  
- ( r ): rente på statsobligasjoner  
- ( g ): vekstrate i BNP  
- ( s ): primærbalanse (skatter – utgifter, ekskl. renter)

Vi antar at ( r, g, s ) er konstante over tid.


(a)

Gitt:

$$
r = 0.03, \quad g = 0.02, \quad s = 0.005
$$

Steady state (langsiktig likevekt):

$$
\dot{b} = 0 \implies (r - g)b_{ss} = s
$$

Dermed:

$$
b_{ss} = \frac{s}{r - g} = \frac{0.005}{0.01} = 0.5 = 50\%
$$

Stabilitet:

Når \( r > g \), får vi en **asymptotisk stabil** likevekt.  
Dette betyr at uansett startverdi vil \( b(t) \) konvergere mot \( b_{ss} \).

**Tolkning:**
- Hvis \( b(0) = 70\% \): \( b(t) \) vil gradvis **falle mot 50\%**.  
- Hvis \( b(0) = 55\% \): \( b(t) \) vil også **falle litt mot 50\%**.

Begge tilfeller oppfyller Maastricht-regelen om at \( b(t) \leq 60\% \) i det lange løp.

---

### (b)

Gitt:

$$
r = 0.01, \quad g = 0.03, \quad s = -0.016
$$

**Steady state:**

$$
b_{ss} = \frac{s}{r - g} = \frac{-0.016}{0.01 - 0.03} = \frac{-0.016}{-0.02} = 0.8 = 80\%
$$

**Stabilitet:**

Når \( r < g \), blir likevekten **ustabil**, men gjelden vil faktisk **falle over tid** fordi økonomien vokser raskere enn rentene.

**Tolkning:**
Selv om \( b_{ss} = 80\% \), vil gjeldsgraden \( b(t) \) faktisk synke mot 0 % uten behov for overskudd.  
Dermed **oppfyller landet 60 %-regelen** på lang sikt, selv uten finanspolitisk inngripen.

## Oppgave 2 – Managing the Stock of a Renewable Resource

Vi betrakter differensialligningen:

$$
\dot{p}(t) = r(p(t)) - (\mu + nk)p(t)
$$

hvor  
- \( p(t) \): fiskebestand  
- \( r(p(t)) = \sqrt{p(t)} \): naturlig reproduksjon  
- \( \mu > 0 \): naturlig dødsrate  
- \( n \): antall fiskere  
- \( k \): sannsynlighet for å fange en fisk

---

### (a)

Steady state finnes ved \( \dot{p} = 0 \):

$$
0 = \sqrt{p_{ss}} - (\mu + nk)p_{ss}
$$

$$
\sqrt{p_{ss}} = (\mu + nk)p_{ss}
$$

$$
1 = (\mu + nk)\sqrt{p_{ss}}
$$

$$
\boxed{p_{ss} = \frac{1}{(\mu + nk)^2}}
$$

---

### (b) Stabilitet

La \( f(p) = \sqrt{p} - (\mu + nk)p \).  
Da er den deriverte:

$$
f'(p) = \frac{1}{2\sqrt{p}} - (\mu + nk)
$$

Ved steady state:

$$
\frac{1}{\sqrt{p_{ss}}} = (\mu + nk)
$$

Sett dette inn:

$$
f'(p_{ss}) = \frac{1}{2}(\mu + nk) - (\mu + nk) = -\frac{1}{2}(\mu + nk) < 0
$$

Siden den er negativ, er steady state **asymptotisk stabil**.

---

### (c)

Gitt:

$$
\mu = 0.03, \quad n = 20, \quad k = 0.001
$$

$$
\mu + nk = 0.03 + 20(0.001) = 0.05
$$

$$
p_{ss} = \frac{1}{0.05^2} = 400
$$

**Steady state fiskebestand:**

$$
\boxed{p_{ss} = 400}
$$

---

### (d)

Nå øker dødsraten og fisket opphører:

$$
\mu = 0.06, \quad n = 0
$$

$$
\mu + nk = 0.06
$$

$$
p_{ss} = \frac{1}{0.06^2} = 277.78
$$

**Sammenligning:**

| Scenario | \( \mu \) | \( n \) | \( p_{ss} \) |
|-----------|-----------|---------|---------------|
| (c) | 0.03 | 20 | 400 |
| (d) | 0.06 | 0 | 277.8 |

**Tolkning:**

Selv om fisket fjernes, reduseres bestand
